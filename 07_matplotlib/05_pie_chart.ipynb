**PIE CHART**  
Categorical Data   
Use case : contribution on standard scale 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
#simple data 
data = [23,45,100,20,49]
plt.pie(data)
plt.show()
```

```python
# adding labels to chart


data = [23,45,100,20,49]
subjects = ['english', 'hindi', 'maths', 'science', 'sst']

plt.pie(data, labels = subjects)
plt.show()
```

```python
# runs in particular match
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'])
plt.show()
```

```python
# auto percentage => autopct()

plt.pie(df['batsman_runs'], labels = df['batsman'], autopct='%0.1f%%') # wtf ??
plt.show()
```

```python
# colors 
plt.pie(df['batsman_runs'], labels = df['batsman'], colors = ['blue','green','yellow','pink','brown','purple'], autopct = '%0.1f%%')
plt.show()
```

```python
# explode

plt.pie(df['batsman_runs'], labels = df['batsman'], autopct='%0.1f%%', explode = [0.1,0,0,0,0,0]) 
plt.show()
```

```python
# shadow 3d feel
plt.pie(df['batsman_runs'], labels = df['batsman'], shadow= True) 
plt.show()
```

```python
#changing styles 

plt.style.available
# these are all available styles
```

```python
plt.style.use('fivethirtyeight')
```
