```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
# Virat Kohli track record of IPL
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# Virat Kohli and Rohit Sharma track record of IPL
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# labels title 
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title("Virat vs Rohit IPL Career")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# colors 

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000000')

plt.title("Virat vs Rohit IPL Career")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")

```

```python
# line style

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'solid') #solid, dashed, dotted, dashdot
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000000', linestyle = 'dashed')

plt.title("Virat vs Rohit IPL Career")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
#line width


plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linewidth = 4)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000000', linestyle = 'solid')

plt.title("Virat vs Rohit IPL Career")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
#marker 


plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', marker = 'o', markersize=10) 
# D, o, +  
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000000', linestyle = 'dashdot', marker = '+')

plt.title("Virat vs Rohit IPL Career")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
#legend

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', marker= 'D', label= 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000000', linestyle = 'dashdot', label='Rohit')

# To print these label on screen we use legend
plt.legend()

# you can also use loc='best' inside legend to set the best location
# plt.legend(loc='best') # upper right, right, left 

plt.title("Virat vs Rohit IPL Career")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# limiting axes
price = [47000, 58000, 51000, 60000, 65000, 48000, 4500000] #an outlier
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)
```

```python
# limiting axes
price = [47000, 58000, 51000, 60000, 65000, 48000, 4500000] #an outlier
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)
plt.ylim(0, 100000) # this tells to show only between this range
```

```python
# limiting axes
price = [47000, 58000, 51000, 60000, 65000, 48000, 4500000] #an outlier
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)
plt.ylim(0, 100000) 
plt.xlim(2017, 2020) # show for only year between 2017 and 2020(including)
```

```python
# grid
# limiting axes
price = [47000, 58000, 51000, 60000, 65000, 48000, 45000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)
plt.grid()
```
