Use case : Frequency count

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
# works on 1 column at a time => numerical

# simple data

data= [ 32, 45, 27, 61, 15, 71]
plt.hist(data)
```

```python
# now i provide my own bin size
data= [ 32, 45, 27, 61, 15, 71]
plt.hist(data, bins = [10,25,40,55,70], edgecolor = 'black')
```

```python
# better bin sizing
data= [ 32, 45, 27, 61, 15, 71]
plt.hist(data, bins=[10, 20, 30, 40, 50, 60, 70, 80], rwidth=0.7)

# rwidth ranges from 0 to 1 and lower the rwidth bigger the gap
# histogram doesn't have gaps in genral but you can for readability using rwidth
```

```python
arr = np.load('big-array.npy')
arr
```

```python
plt.hist(arr, bins = [10,20,30,40,50,60,70,80,90])
plt.show()
# here we can't see values which has very less frequency
```
