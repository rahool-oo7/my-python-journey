##### INDIAN EMPLOYEE DATA

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('indian_employee_data.csv')

df.head()
```

```python
# checking number of missing values of each column
print("missing values in each column")

print(df.isnull().sum())
```

```python
print(df.isin([np.inf, -np.inf]).sum())


```

```python
# it's best to convert the infinite value as nan to correct both using fillna()
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# handling salary missing data

# avg of salary excluding missing value 

avg_salary = df['Salary'].mean()
# Pandas .mean() function ignores NaN by default.

# df['Salary'].fillna(avg_salary, inplace=True) # NOT RELIABE IN PANDAS 3.O

df['Salary'] = df['Salary'].fillna(avg_salary)
df.head()

```

```python
# handling experience

exp = round(df['Experience'].mean())

df['Experience'] = df['Experience'].fillna(exp)
df.head()
```

```python
def count_vowel(words):

    count = 0 
    vowel = 'aeiou'
    for i in words.lower() :
        if i in vowel:
            count+=1
    
    return count

str1  = input("Enter a string : ")
print("No. of vowels: ",count_vowel(str1))
```
