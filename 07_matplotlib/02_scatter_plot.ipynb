**SCATTER PLOT**

Bivariate analysis  
Works mostly on Numberical and Numberical column  
Use case - Find co-relation b/w two quantities




```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
# plt.scatter simple function

x = np.linspace(-10, 10, 50) # 50 equally disatant points b/w -10 and 10

y = 10*x + 3 + np.random.randint(1,300,50)

plt.scatter(x, y)
```

```python
df = pd.read_csv('batter.csv')
df
```

```python
# take top 50 batsman
df = df.head(50)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'] )
plt.title("Avg and SR analysis of top 50 batsman")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
```

```python
# ploting the same using plt.plot() => useful for Larger dataset
# This is faster but no scatter specific functions can be used 
# scatter is slower but gives various parameters to work on


plt.plot(df['avg'], df['strike_rate'], 'o')
```
