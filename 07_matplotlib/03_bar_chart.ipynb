**BAR CHART**    
1 Categorical data and 1 Numerical data    
eg. No of children(numerical) liking specific color(categorical)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
# simple bar chart 

children = [20, 30,12, 18, 24]

colors = ['red', 'yellow', 'blue', 'green', 'orange']

plt.bar(colors, children, color = 'purple')
plt.title("Children's Favourite colour")
plt.xlabel('Colours')
plt.ylabel("No. of Children")

# This is vertical bar chart 
```

```python
# simple bar chart 

children = [20, 30,12, 18, 24]

colors = ['red', 'yellow', 'blue', 'green', 'orange']

plt.barh(colors, children, color = 'green')
plt.title("Children's Favourite colour")
plt.ylabel('Colours')
plt.xlabel("No. of Children")

# This is horizontal bar chart => don't forget to change x and y Label
# just use barh instead of bar
```

```python
# 3 year data on top 5 batsman

df = pd.read_csv('batsman_season_record.csv')
df
```

```python
# runs scored by each in 2015
plt.bar(df['batsman'], df['2015'])
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])

# this displays 3 bar graph but OVERLAPING therefore not GOOD
# We need to place bars side by side, 3 BARS for each player
```

```python
# problem
children = [20, 30,12, 18, 24]

colors = ['red red red red', 'yellow yellow yellow yellow', 'blue blue blue blue', 'green reen reen reen', 'orangeorangeorangeorange']

# color name will overlap
plt.bar(colors, children)
```

```python
# fixing using xticks()
children = [20, 30,12, 18, 24]

colors = ['red red red red', 'yellow yellow yellow yellow', 'blue blue blue blue', 'green reen reen reen', 'orangeorangeorangeorange']

# color name will overlap
plt.bar(colors, children)
# plt.xticks(rotation=70)
plt.xticks(rotation='vertical')
```

```python
# stacked bar chart 
# total runs of each player with distinct years in different color

plt.bar(df['batsman'], df['2015'], label = '2015')
plt.bar(df['batsman'], df['2016'], bottom= df['2015'], label = '2016') # adding after 2015 length
plt.bar(df['batsman'], df['2017'], bottom= df['2015'] + df['2016'], label = '2017') 
# adding after length of 2015 + 2016 combined
# if you do here only 2016 it will plot after length of 2016 bar, which will overlap
# In stacking, each next layer must start where the total height of previous layers ends (cumulative sum)


plt.legend()

```

```python
plt.bar(df['batsman'], df['2017'], label = '2017')
plt.bar(df['batsman'], df['2016'], bottom= df['2017'], label = '2016')
plt.bar(df['batsman'], df['2015'], bottom= df['2017'] + df['2016'], label = '2015') 
plt.legend()
```

```python
#multiple graphs

x = np.arange(len(df['batsman']))  # number of things to show in x axis
x_val = x = np.arange(len(df['batsman'])) # for future changing back to name in x axis
w = 0.25 # width of bars
plt.bar(x, df['2015'], width = w, label = '2015')

x = x+w         # to add other bar after that length
plt.bar(x, df['2016'], width = w, label = '2016')

x = x+w
plt.bar(x, df['2017'], width = w, label = '2017')

plt.xticks((x_val + w), df['batsman']) # changing numerical value to names of batsman
# + w to adjust the x ticks in center of three bars

plt.legend(loc = 'best')
plt.show()
```
